```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

# Load your dataset
# Replace 'your_dataset.csv' with the actual file name and path
path = "/content/sample_data/Mall_Customers.csv"
data = pd.read_csv(path)
data
```

```python
# Display the first few rows of the dataset to understand its structure
print(data.head())
```

```python
#Modify this based on your dataset
X = data.iloc[:, [3, 4]].values
```

```python
# Standardize the features to ensure they have the same scale
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
# Function to plot the Elbow Method graph
def plot_elbow_method(X, max_clusters=10):
    wcss = []  # Within-Cluster-Sum-of-Squares

    for i in range(1, max_clusters + 1):
        kmeans = KMeans(n_clusters=i, init='k-means++', random_state=42)
        kmeans.fit(X)
        wcss.append(kmeans.inertia_)

    # Plot the Elbow Method graph
    plt.plot(range(1, max_clusters + 1), wcss, marker='o')
    plt.title('Elbow Method')
    plt.xlabel('Number of Clusters')
    plt.ylabel('WCSS')  # Within-Cluster-Sum-of-Squares
    plt.show()

# Call the function with your scaled dataset
plot_elbow_method(X_scaled, max_clusters=10)
```

```python
# Visualize the clusters
plt.scatter(X_scaled[:, 0], X_scaled[:, 1], c=kmeans.labels_, cmap='viridis')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=300, c='red', marker='X')
plt.title('K-Means Clustering')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.show()
```
